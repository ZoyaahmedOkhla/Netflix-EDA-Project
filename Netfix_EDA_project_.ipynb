```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
# clean data
df=df.dropna(subset=["type","release_year","rating","country","duration"])
# how many movies vs tv shows?=> Bar chart
type_counts=df["type"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=["orange","pink"])
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.title("distribution of movie & tv shows")
plt.savefig("type_distribution.png")
plt.show()
```

```python
#2)what is the percentage of each content rating=> pie chart
rating_counts=df["rating"].value_counts()
plt.figure(figsize=(10,8))
plt.pie(rating_counts,labels=rating_counts.index,autopct="1%.1f%%",startangle=90)
plt.title("content rating percentage")
plt.tight_layout()
plt.savefig("content_rating_pie.png")
plt.show()
```

```python
# what is the distribution of movie duration??=>Histrogram
movie_duration=df[df["type"]=="Movie"].copy()
movie_duration['duration']=movie_duration['duration'].str.replace('min','').astype(int)

plt.figure(figsize=(8,6,))
plt.hist(movie_duration['duration'],bins=20,color='skyblue',edgecolor='black')
plt.xlabel(" Duration(min)")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Duration")
plt.tight_layout()
plt.savefig("movie_duration_hist.png")
plt.show()
```

```python
# relation between release year and number of shows=> scatter plot
release_counts=df["release_year"].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color="red",marker="o")
plt.xlabel("Release Year")
plt.ylabel("Number of Shows")
plt.title("Relation between release year and number of shows")
plt.tight_layout()
plt.savefig("release_year_scatter.png")
plt.show()
```

```python
# top 10 countries with highest number of shows ??=> Bar chart
country_counts=df["country"].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.bar(country_counts.index,country_counts.values,color="pink")
plt.xlabel("numbers of shows")
plt.ylabel("country")
plt.title("Top 10 Countries with Highest Number of Shows")
plt.tight_layout()
plt.savefig("top_10_countries.png")
plt.show()
```

```python
# compare multiple plots together ()=> subplots
# group by
content_by_year=df.groupby(["release_year","type"]).size().unstack().fillna(0)
fig,axs=plt.subplots(1,2,figsize=(10,6))
# first subplot :movies
axs[0].plot(content_by_year.index,content_by_year["Movie"],marker="o",linestyle="-",color="red",label="Movies")
axs[0].set_title("Movies released over the year")
axs[0].set_xlabel("year")
axs[0].set_ylabel("number of shows")
axs[0].legend()

# second subplot: TV shows
axs[1].plot(content_by_year.index, content_by_year["TV Show"], marker="o", linestyle="-", color="orange", label="TV Shows")
axs[1].set_title("TV Shows released over the year")
axs[1].set_xlabel("year")
axs[1].set_ylabel("number of shows")
axs[1].legend()


fig.suptitle("comparison  of movies and tv shows released over years")
plt.tight_layout()
plt.savefig("comparison_subplots.png")
plt.show()
```
